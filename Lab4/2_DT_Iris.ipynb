```python
#Import scikit-learn dataset library
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import numpy as np

#Load dataset
iris = datasets.load_iris()
```

```python
# print the names of the 13 features
print("Features:", iris.feature_names)

# print the label type of wine(class_0, class_1, class_2)
print("\nLabels:", iris.target_names)

# print data(feature)shape
iris.data.shape
```

```python
#import the necessary module
from sklearn.model_selection import train_test_split

#split data set into train and test sets
train_x, test_x, train_y, test_y = train_test_split(iris.data, iris.target, test_size = 0.3, random_state = 10)
```

```python
#Create a Decision Tree Classifier (using Gini)
clf = DecisionTreeClassifier(criterion = "gini")

#Train the model using the training sets
final_model = clf.fit(train_x, train_y)
```

```python
# Predict the classes of test data
prediction = clf.predict(test_x)
print("prediction:",prediction)
print(prediction.dtype)
```

```python
from sklearn import metrics

# Model Accuracy, how often is the classifier correct?
Accuracy = metrics.accuracy_score(test_y, prediction)
print("Accuracy:",Accuracy)
```

```python
from sklearn.tree import export_graphviz
export_graphviz(clf,out_file='iris_tree.dot',feature_names=list(iris.feature_names),
               class_names=list(iris.target_names), filled=True)

# Convert to png
from subprocess import call
call(['dot', '-Tpng', 'iris_tree.dot', '-o', 'iris_tree.png', '-Gdpi=600'])
```

```python
# Display in python
import matplotlib.pyplot as plt
plt.figure(figsize = (14, 18))
plt.imshow(plt.imread('iris_tree.png'))
plt.axis('off')
plt.show()
```
